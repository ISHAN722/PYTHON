LINKED LIST


1.   create Node class
2.   create linkedlist class
3.   insert elements using insert()
4.   print data using printlist()
5.   delete elements using delete()
6    reverse the LL using reverse()
7.   find the length of the LL using length()


```python
class Node:

  def __init__(self,data):
    self.data = data
    self.next = None

class LinkedList:
  
  def __init__(self):
    self.head = None

  def insert(self, index, new_value):
    new_node = Node(new_value)
    if self.head is None:
      new_node = self.head 
    elif index == 0:   
      new_node.next = self.head
      self.head = new_node 
    else:
      last = self.head
      for i in range(0, index-1):
        last = last.next
      new_node.next = last.next
      last.next = new_node   

  def delete(self, index):
    t = self.head
    for i in range(0, index-1):
      t = t.next
    t.next = t.next.next  

  def reverse(self): 
    t = self.head
    p = Node(None)
    q = Node(None)
    while(t):
      q = p
      p = t
      t = t.next
      p.next = q  
    self.head = p    

  def length(self):
    t = self.head
    l = 0
    while(t):
      l = l + 1
      t = t.next
    print(l)    

  def printlist(self):
    t = self.head
    while(t):
      if(t.next == None):
        print(t.data, end = " ")
      else:
        print(t.data, end = "->")  
      t = t.next 

if __name__ == '__main__':
  arr = list(map(int,input().split()))
  newlist = LinkedList()
  newlist.head = Node(arr[0])
  last = newlist.head
  for i in range(1,len(arr)):
    t = Node(arr[i])
    t.next = None
    last.next = t
    last = t
  newlist.delete(3)
  newlist.printlist()


```

```python
class Node:

  def __init__(self,data):
    self.data = data
    self.next = None

class Circularlinkedlist:

  def __init__(self):
    self.head = None

  def insert(self, index, value):
    new_node = Node(value)
    if self.head is None:
      self.head = new_node
    elif index == 0:  
      last = self.head 
      while last.next != self.head:
         last = last.next
      last.next = new_node
      new_node.next = self.head
      self.head = new_node
    else:
      last = self.head
      for i in range(0, index-1):
        last = last.next
      new_node.next = last.next
      last.next = new_node  

  def count(self):
    t = self.head
    l = 0
    while True:
      l = l + 1
      t=t.next
      if(t == self.head):
        break
    return l    
  
  def delete(self,index):
     t = self.head
     if (self.head == None):
       print("empty list")  
     if (index < 0 or self.count() < index):
       print("invalid index")    
     elif (index == 1):
        while(t.next != self.head):
           t = t.next
     elif (t == self.head):
       self.head = None  
       self.head = t.next
     else:
       for i in range(0, index-1):
          t = t.next
          t.next = t.next.next  
  
  def reverse(self):
    t = self.head
    p = Node(None)
    q = Node(None)
    while True:
      q=p
      p=t
      t=t.next
      p.next=q
      if(t == self.head):
        break
    self.head.next = p
    self.head = p      
  
  def printlist(self):
    t = self.head
    while True:
      print(t.data, end=" ")
      t = t.next
      if(t == self.head):
        break                                                                     

if __name__ == "__main__":
  arr = list(map(int, input().split()))
  newlist = Circularlinkedlist()
  newlist.head = Node(arr[0])
  newlist.head.next = newlist.head
  last = newlist.head
  for i in range(1, len(arr)):
    t = Node(arr[i])
    t.next = last.next
    last.next = t
    last = t
  newlist.delete(4)  
  newlist.printlist()
  
  
  

```

STACK


1.   maxsize – Number of items allowed in the queue if not mentioned then size is infinite
2.   empty() – Return True if the queue is empty, False otherwise
3.   full() – Return True if there are maxsize items in the queue
4.   get() – Remove and return an item from the queue. If queue is empty, wait until an item is available
5.   put(item) – Put an item into the queue. If the queue is full, wait until a free slot is available before adding the item.

6.   qsize() - return number of elements in stack

```python
from queue import LifoQueue
 
stk = LifoQueue(maxsize = 0)
 
print(stk.qsize())
  
stk.put(1)
stk.put(2)
stk.put(3)
stk.put(4)
stk.put(5)
stk.put(6)

print("Is the stack Full: ", stk.full())
print("The size is: ", stk.qsize())
print(stk.get())
print(stk.get())
print(stk.get()) 
print("Is the stack Empty: ", stk.empty())
 
```

USING QUEUE MODULE


1.   from queue module import PriorityQueue
2.   create priority queue using --PriorityQueue
3.   add items using --put()
4.   remove items using --get()

empty(), full(), qsize() are not reliable as they risk race condition where the queue size might change.

```python
from queue import PriorityQueue

q = PriorityQueue()
q.put((6,'a'))
q.put((3,'b'))
q.put((5,'c'))
q.put((1,'d'))
q.put((2,'e'))
q.put((4,'f'))

print(q.get())
print(q.get())

print("number of items in PriorityQueue: ",q.qsize())
print("is the PriorityQueue empty: ",q.empty())
print("is the PriorityQueue full: ",q.full())
```

BINARY SEACRH TREE

1.  create a Node class with left,right and value


```python
class Node:
  def __init__(self,data):
    self.left = None
    self.right = None
    self.data = data

def search(root,data):
  if root is None or root.data == data:
    return root
  elif root.data < data:
    return search(root.right, data)  
  else:
    return search(root.left, data)

def insert(root,data):
  if root == None:
    return Node(data)
  else:
    if root.data < data:
      root.right = insert(root.right, data)
    else:
      root.left = insert(root.left,data)
    return root    

def inorder(root):
  if root:
    inorder(root.left)
    print(root.data, end = " ")
    inorder(root.right) 

def postorder(root):
  if root:
    postorder(root.left)
    postorder(root.right)
    print(root.data, end = " ")

def preorder(root):
  if root:
    print(root.data, end = " ")
    preorder(root.left)
    preorder(root.right)   

def minValueNode(node):
  current = node
  while current.left is not None:
    current = current.left
  return current 

def delete(root, data):
  if root is None:
    return root
  elif root.data < data:
    root.right = delete(root.right, data)
  elif root.data > data:
    root.left = delete(root.left, data)
  else:
    if root.left is None:      # if the current node has only one or zero child nodes
      temp = root.right        
      root = None
      return temp
    elif root.right is None:
      temp = root.left
      root = None
      return temp

    temp = minValueNode(root.right)    # if the current node has 2 child nodes
    root.data = temp.data
    root.right = delete(root.right,temp.data)
  return root


root = None
root = insert(root, 50)
root = insert(root, 30)
root = insert(root, 20)
root = insert(root, 40)
root = insert(root, 70)
root = insert(root, 60)
root = insert(root, 80)


inorder(root)
print(" ")
root = delete(root,50)
inorder(root)
```

HEAP QUEUE

it is mainly used to represent a priority queue and can be used using the heapq module


1.   import heapq and initialize a list
2.   convert the list into a heap using heapq.heapify()
3.   elements can be inserted using heapq.heappush()
4.   elements can be deleted using heapq.heappop()
5.   elements can and be pushed and popped simultaneously using heapq.heappushpop()
6.   element is popped and the pushed using heapq.heapreplace()
7.   return the kth largest element using heapq.nlargest()
8.   return the kth smallest element using heapq.nsmallest() 


```python
import heapq

print("FIRST EXAMPLE")
arr1 = [1,8,4,6,0,2,5,78,9,3]
heapq.heapify(arr1)
print("the heap is: ",list(arr1))
heapq.heappush(arr1, 7)
print("the heap is: ",list(arr1))
print("element popped: ",heapq.heappop(arr1))
print("the 3 largest numbers are: ", heapq.nlargest(3, arr1))
print("the 3 smallest numbers are: ", heapq.nsmallest(3, arr1))
print("pushing and popping an element: ", heapq.heappushpop(arr1, -1))
print("popping and pushing an element: ", heapq.heapreplace(arr1, -1))
print("SECOND EXAMPLE")
arr2 = [9,6,4,67,9,3,56,8,0,3]
heapq.heapify(arr2)
print("the heap is: ",list(arr2))
heapq.heappush(arr2, 2)
print("the new heap is: ",list(arr2))
print("element popped is: ",heapq.heappop(arr2))
print("the 3 largest numbers are: ",heapq.nlargest(4, arr2))
print("the 3 smallest numbers are: ",heapq.nsmallest(4, arr2))
print("pushing and popping an element: ", heapq.heappushpop(arr2, -1))
print("popping and pushing an element: ", heapq.heapreplace(arr2, -1))
```
